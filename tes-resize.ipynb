```python
import cv2
import matplotlib.pyplot as plt
import os
import random
```

```python
def resize_and_show(image_path, target_size):
    # Membaca gambar menggunakan OpenCV
    image = cv2.imread(image_path)
    original_size = image.shape[:2]

    # Mengubah ukuran gambar
    resized_image = cv2.resize(image, target_size)
    
    # Menampilkan gambar sebelum dan setelah diresize
    plt.subplot(1, 2, 1)
    plt.title(f"Ori {original_size[0]}x{original_size[1]}")
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title(f"After {target_size[0]}x{target_size[1]}")
    plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.show()


def pick_random_image(directory):
    files = os.listdir(directory)
    
    random_image = random.choice(files)
    image_path = os.path.join(directory, random_image)

    return image_path

```

```python
if __name__ == "__main__":
    cnn_arab_input = (32, 32)
    vgg_input = (224, 224)
    inception_resnetv2 = (229, 229)
    target_sizes = [cnn_arab_input, vgg_input, inception_resnetv2]
    
    corrected_path = './Dyslexia Dataset/Train/Corrected/'
    normal_path = './Dyslexia Dataset/Train/Normal/'
    reversal_path = './Dyslexia Dataset/Train/Reversal/'
    target_paths = [corrected_path, normal_path, reversal_path]

    for path in target_paths:
        print(f'==> {path.split("/")[-2]} <==')
        # print(path)
        for target_size in target_sizes:
            image_path = pick_random_image(path)
            resize_and_show(image_path, target_size)
```
